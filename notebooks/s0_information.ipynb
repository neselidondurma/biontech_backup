```python
import os
import matplotlib.pyplot as plt
import pandas as pd
from PIL import Image
```

```python
def display_image(img_fpath: str, target_width: int = 900):
    """
    Display an image with a consistent width while maintaining the aspect ratio.

    Parameters:
        img_fpath (str): Path to the image file.
        target_width (int): The desired width for the image (default is 300 pixels).
    """
    img = Image.open(img_fpath)
    w, h = img.size

    # Calculate the new height to maintain the aspect ratio
    aspect_ratio = h / w
    target_height = int(target_width * aspect_ratio)

    # Resize the image
    img = img.resize((target_width, target_height))
    display(img)
```

- Pancreatic ductal adenocarcinoma (PDAC) is lethal in 88% of patients but mutation-derived T cell neoantigens that are suitable for vaccines
- we synthesized mRNA neoantigen vaccines in real time from surgically resected PDAC tumours
- treated 16 patients with atezolizumab and autogene cevumeran, then 15 patients with mFOLFIRINOX
- PDACs are almost completely insensitive (<5% response rate11,12) to
immune checkpoint inhibitors.
-most PDACs in fact harbour more neoantigens
than previously predicted1
- long-term survivors of PDAC mount
spontaneous T cell responses against tumour-specific neoantigens not
shared among patients --> a general treatment for all is not realisable

Hypothesis: an effective
individualized mRNA vaccine would induce neoantigen-specific
T cells in PDAC, eliminate micrometastases and delay recurrence.

# Workflow:


1) Identify Neoantigens: 
Sequence Tumor per patient, find 20 neoantigens
2) Vaccine Design: 
Encoding neoantigens into mRNA molecules, targeting both MHC1 and 2, recognized by CD8 and CD4+
3) Packagging in Lipoplex Nanoparticles: 
Goal: Protection from degradation in blood, delivery to immune cell via i.v. --> we want mRNA to reach to immune cells such as dendritic cells

Role of chemoterapy here: altering the tumor microenvironment, inducing immunogenic cell death =releasing tumor antigens and enhancing immune response

2 Goals of vaccine:
1) Amplifying Existing T-Cells: by presenting neoantigens and combining this with immune checkpoint inhibitors
2) Prime T Cells: Creating e new population of activated T-Cells that can specifically target tumor

# 21 CFR Part 11

is a regulation issued by the U.S. Food and Drug Administration (FDA) to determine which electronic records and electronic signatures are considered trustworthy, reliable, and equivalent to paper records and handwritten signatures.

1) System Validation:

Systems must be validated to ensure accurate, consistent, and reliable operation.

2) Audit Trials: 

Computerized systems must have secure, time-stamped audit trails that document actions such as data creation, modification, or deletion.

3) Protection of Records:

protected from unauthorized access, tampering, or loss + backup

4) Electronic Signatures:

Should be unique, can be biometric or non-biometric(password based), should be linked to electornic records, so they can not be changed anonymously

5) SOPs:

Standard Operating Procedures (SOPs): Companies must establish and maintain SOPs that outline how systems are managed to comply with Part 11 + periodic review
